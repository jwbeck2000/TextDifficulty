```
from wordcloud import WordCloud
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```
data = np.load('/work/x_train.npy', allow_pickle = True).astype(str)
topics_df = pd.read_csv('/work/df_document_topic (1).csv')

topics_df.head()
```

```
data_df = pd.DataFrame(data,columns = ['Text'])
data_df.head()
```

```
comb_df = pd.concat([data_df,topics_df],axis=1)
comb_df.head()
```

```
# list = comb_df.Text.tolist()
# new_list = [i.split() for i in list]
# final_list = [item for sublist in new_list for item in sublist]
```

```
clouds = []
for i in range(0,5):
    df = None
    list = None
    new_list = None
    final_list = None 

    df = comb_df[comb_df['dominant_topic']==i]
    list = df['Text'].tolist()
    new_list = [i.split() for i in list]
    final_list = [item for sublist in new_list for item in sublist]
    final_list = ','.join(final_list)

    wordcloud = WordCloud(background_color = 'white', contour_width = 2)
    viz = wordcloud.generate(final_list)
    clouds.append(viz)
    
```

```
clouds
```

```
for i in clouds:
    plt.figure(figsize = (6, 6), facecolor = None)
    plt.imshow(i)
    plt.axis("off")
    plt.tight_layout(pad = 0)
    
    plt.show()
```
